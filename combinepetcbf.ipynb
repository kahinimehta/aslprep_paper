```python
nilist = pd.read_csv('NKI_list.txt')
nkipet = pd.read_csv('scripts/nkidata/NKI_cbfpetcorr.csv')
nk = nilist.join(nkipet)
nk3 = pd.read_csv('scripts/nkidata/NKI_participants_New.tsv',sep='\t')
nk = pd.merge(nk,nk3,on=["sub","ses"])

pn1 = pd.read_csv('scripts/pncdata/PNC_cbfpetcorr.csv')
pn1= pn1.drop_duplicates()
pn2 = pd.read_csv('scripts/pncdata/pnc_demographic.csv')
pn2 = pn2.drop_duplicates()
pn = pd.merge(pn1,pn2,on="sub")

gr1=pd.read_csv('scripts/grympydata/demographics.csv')
gr2 =pd.read_csv('scripts/grympydata/IRR_cbfpetcorr.csv')
gr = pd.merge(gr1,gr2,on="sub")



ags1 =pd.read_excel('scripts/ageingdata//Book1.xlsx',engine='openpyxl')
ags2 =pd.read_csv('scripts/ageingdata/AGE_cbfpetcorr.csv')
ag = pd.merge(ags1,ags2,on='sub')

cd1 = pd.read_excel('scripts/ftddata/Updated_LEFFTDS.xlsx',engine='openpyxl')
cd2 = pd.read_csv('scripts/ftddata/FTD_cbfpetcorr.csv')
cd = pd.merge(cd1,cd2,on='sub')
```

```python
# combine all the data and label with common names 
agdata = pd.DataFrame({'sub':ag['sub'],'Datasets': np.repeat('AGE',len(ag)),'SCBF': ag['SCBF'],'AGE':ag['age'],
                       'BASIL': ag['BASIL'],'SCRUB': ag['SCRUB'],'PVCGM': ag['PVCGM'],'PVCWM': ag['PVCWM']})


grdata  = pd.DataFrame({'sub':gr['sub'],'Datasets': np.repeat('IRR',len(gr)),'SCBF': gr['SCBF'],'AGE':gr['age'],'BASIL': gr['BASIL'],
                       'SCRUB': gr['SCRUB'],'PVCGM': gr['PVCGM'],'PVCWM': gr['PVCWM']})

cddata = pd.DataFrame({'sub':cd['sub'],'Datasets': np.repeat('FTD',len(cd)),'SCBF': cd['SCBF'],'AGE':cd['age'],'BASIL': cd['BASIL']
                        ,'PVCGM': cd['PVCGM'],'PVCWM': cd['PVCWM']})



pndata  = pd.DataFrame({'sub':pn['sub'],'Datasets': np.repeat('PNC',len(pn)),'SCBF': pn['SCBF'],'AGE':pn['age'],
                       'BASIL': pn['BASIL'],'SCRUB': pn['SCRUB'],'PVCGM': pn['PVCGM'],'PVCWM': pn['PVCWM'], })



nkdata1 = pd.DataFrame({'sub':nk['sub']+nk['ses'],'Datasets': np.repeat('NKI',len(nk)),'SCBF': nk['SCBF'],'AGE': nk['age'],'BASIL': nk['BASIL'],
                   'SCRUB': nk['SCRUB'],'PVCGM': nk['PVCGM'],'PVCWM': nk['PVCWM']   })

gmdata = pd.concat([pndata,nkdata,grdata,cddata,agdata])
```

```python
gmdata.to_csv('petcbfcorr.csv',index=False,sep=',')
```
